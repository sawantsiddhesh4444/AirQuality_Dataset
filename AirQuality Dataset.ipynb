```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Convert 'Jan' column to numeric
df['Jan'] = pd.to_numeric(df['Jan'], errors='coerce')

# Ensure 'Country' column is categorical
df['Country'] = df['Country'].astype('category')

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Jan', y='Country', data=df)
plt.title('AQI Values in January 2023 for Different Cities')
plt.xlabel('AQI Value')
plt.ylabel('Country')
plt.show()
```

```python
# Plot histogram for the 'Jan' column
plt.figure(figsize=(10, 6))
plt.hist(df['Jan'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of AQI Values in January 2023')
plt.xlabel('AQI Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot the bar plot
sns.barplot(x='Jan', y='City', data=df, palette='viridis')

# Set the title and labels
plt.title('AQI Values in January 2023 for Different Cities', fontsize=16)
plt.xlabel('AQI Value', fontsize=14)
plt.ylabel('City', fontsize=14)
plt.show()
```

```python
# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Jan', y='Country', data=df)
plt.title('Scatter Plot of AQI Values in January 2023 for Different Cities')
plt.xlabel('AQI Value')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Rank', y='Jan', data=df)
plt.title('Line Plot of AQI Values in January 2023 for Different Cities')
plt.xlabel('Rank')
plt.ylabel('AQI Value')
plt.show()
```
